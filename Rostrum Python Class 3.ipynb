```python
class conics:

  pi = 3.1416

  def __init__(self , a,b,h,f,g):
    self.a = a
    self.b = b
    self.h = h
    self.f = f
    self.g = g

  def get_type(self):
    discriminant = (self.b ** 2 - 4 * self.a * self.f)
    if discriminant > 0:
      return "ellipse"
    elif discriminant < 0:
      return "hyperbola"
    else:
      return "parabola"

  def area(self):
    if self.get_type() == "ellipse":
      return self.pi * self.a * self.b
    elif self.get_type() == "hyperbola":
      return 2 * self.pi * self.a * self.b
    else:
      return 0.5 * self.a * self.b
```

```python
c = conics(1,2,0.5,1.5,2)
c.get_type()
```

```python
c.area()
```

Inheritance

```python
class Vehicle:
    def __init__(self, name, speed):
        self.name = name
        self.speed = speed

    def move(self):
        print(f"{self.name} is moving at {self.speed} km/h")

class Car(Vehicle):
    def __init__(self, name, speed, num_wheels):
        super().__init__(name, speed)
        self.num_wheels = num_wheels

    def move(self):
        print(f"{self.name} is moving at {self.speed} km/h on {self.num_wheels} wheels")

class Bicycle(Vehicle):
    def __init__(self, name, speed, num_wheels):
        super().__init__(name, speed)
        self.num_wheels = num_wheels

    def move(self):
        print(f"{self.name} is moving at {self.speed} km/h on {self.num_wheels} wheels")

# Create instances of the Vehicle, Car, and Bicycle classes
vehicle1 = Vehicle("Toyota Camry", 120)
car1 = Car("Honda Civic", 100, 4)
bicycle1 = Bicycle("Giant Defy", 20, 2)

# Create a list of vehicles
vehicles = [vehicle1, car1, bicycle1]

# Iterate over the list of vehicles and call the move() method on each object
for vehicle in vehicles:
    vehicle.move()
```

```python
for vehicle in vehicles:
    print(vehicle)
```

```python
class Vehicle:
    def __init__(self, name, speed):
        self.name = name
        self.speed = speed

    def move(self):
        print(f"{self.name} is moving at {self.speed} km/h")

class Car(Vehicle):
    def __init__(self, name, speed, num_wheels):
        super().__init__(name, speed)
        self.num_wheels = num_wheels

    def move(self):
        print(f"{self.name} is moving at {self.speed} km/h on {self.num_wheels} wheels")

    def __str__(self):
      return "Car name "+self.name+", speed " +str(self.speed)

class Bicycle(Vehicle):
    def __init__(self, name, speed, num_wheels):
        super().__init__(name, speed)
        self.num_wheels = num_wheels

    def move(self):
        print(f"{self.name} is moving at {self.speed} km/h on {self.num_wheels} wheels")

    def __str__(self):
      return "Bicycle name "+self.name+", speed " +str(self.speed)
```

```python
vehicle1 = Vehicle("Toyota Camry", 120)
car1 = Car("Honda Civic", 100, 4)
bicycle1 = Bicycle("Giant Defy", 20, 2)

# Create a list of vehicles
vehicles = [vehicle1, car1, bicycle1]

# Iterate over the list of vehicles and call the move() method on each object
for vehicle in vehicles:
    print(vehicle)
```

For this challenge, create a bank account class that has two attributes:

    owner
    balance

and two methods:

    deposit()
    withdraw()

As an added requirement, withdrawals may not exceed the available balance.

Instantiate your class, make several deposits and withdrawals, and test to make sure the account can't be overdrawn.

```python
class bank_account:

  def __init__(self,owner,balance):
    self.owner=owner
    self.balance=balance

  def deposit(self,amount):
    self.balance+=amount
    print("the deposit of "+str(amount)+"has been done")

  def withdraw(self,am):
   if am <= self.balance:
    self.balance -= am
    print("done")
   else :
    print("withdraw not possible")

  def __str__(self):
    return self.owner + "'s account has balance " + str(self.balance)
```

```python
ab = bank_account("Rafia",1000)
ab.deposit(500)
ab.withdraw(1400)
print(ab)
```
