If Else Statement

```python
n = 13

# checking even odd
if n%2 == 0:
  print("even number")
else:
  print("odd number")
```

```python
a , b , c = 5, 9, 30

#finding largest among 3
if a>b and a>c:
  print("The largest number is " , a)
elif b>a and b>c:
  print("The largest number is " , b)
else:
  print("The largest number is " , c)
```

```python
# prompt: show an example of ternary operator using ? sign

a = 5
b = 10
c = 15
d = a if a > b else b if b > c else c
print(d)
```
