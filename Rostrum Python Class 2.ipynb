Break, Continue

```python
x = 10

while x<=100:
  print(x)
  if x>50:
    break
  x+=10
```

```python
x = 10

while x<=100:
  if x == 50:
    x+=10
    continue
  print(x)
  x+=10
```

Nested Loop

```python
for i in range(5):
  for j in range(5):
    print(j , end="")
  print()
```

```python
for j in range(5):
  print(j , end="")
```

```python
for i in range(8):
  for j in range(1, i+1):
    print(j , end="")
  print()
```

```python
def find_list_sum(l):
  sum = 0
  for i in l:
    sum+=i

  return sum
```
